```python
v1 = np.array((1, 2, 3))
v2 = np.array((3, 3, 3))

print(v1.dot(v2))
print(v1 @ v2)
```

```python
v1 = np.array((1, 2, 3))
v2 = np.array((3))

print(v1.dot(v2))
print(v1.dot(v2).sum())
```
